```python
#import libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#load dataset
sentiment = pd.read_csv('fear_greed.csv')
trades = pd.read_csv('hyperliquid.csv')
```

```python
#show first 5 rows of dataset
print(sentiment.head())
```

```python
print(trades.head())
```

```python
sentiment['timestamp'] = pd.to_datetime(sentiment['timestamp'], unit='s')
sentiment.head()
```

```python
trades['Timestamp IST'] = pd.to_datetime(trades['Timestamp IST'], format='%d-%m-%Y %H:%M')
trades.head()
```

```python
sentiment['date_only'] = sentiment['timestamp'].dt.date
sentiment.head()
```

```python
trades['date_only'] = trades['Timestamp IST'].dt.date
trades.head()
```

```python
merged = trades.merge(
    sentiment[['date_only', 'value', 'classification']],
    on='date_only',
    how='left'
)

merged.head()
```

```python
merged.isnull().sum()
```

```python
# Fill missing sentiment values
merged['value'] = merged['value'].fillna(method='ffill').fillna(method='bfill')
merged['classification'] = merged['classification'].fillna(method='ffill').fillna(method='bfill')

merged.isnull().sum()
```

```python
# Convert Side to numeric: BUY = +1, SELL = -1
merged['side_numeric'] = merged['Side'].apply(lambda x: 1 if x.upper() == "BUY" else -1)
```

```python
merged['profit_flag'] = merged['Closed PnL'].apply(lambda x: 1 if x > 0 else 0)
```

```python
merged['Size USD'] = pd.to_numeric(merged['Size USD'], errors='coerce')
```

```python
sentiment_map = {
    'Fear': 0,
    'Extreme Fear': 0,
    'Greed': 1,
    'Extreme Greed': 1
}

merged['sentiment_numeric'] = merged['classification'].map(sentiment_map)
```

```python
merged.head()
```

```python
sns.countplot(x='Side', data=merged)
plt.title("Count of BUY vs SELL trades")
plt.show()
```

```python
sns.countplot(x='profit_flag', data=merged)
plt.title("Number of Profitable vs Losing trades")
plt.xticks([0,1], ['Loss', 'Profit'])
plt.show()
```

```python
sns.boxplot(x='sentiment_numeric', y='Size USD', data=merged)
plt.xticks([0,1], ['Fear', 'Greed'])
plt.title("Trade Volume (USD) under Fear vs Greed")
plt.show()
```

```python
profit_summary = merged.groupby('sentiment_numeric')['profit_flag'].mean().reset_index()
profit_summary['sentiment_label'] = profit_summary['sentiment_numeric'].map({0:'Fear', 1:'Greed'})
profit_summary
```

```python
volume_summary = merged.groupby('sentiment_numeric')['Size USD'].mean().reset_index()
volume_summary['sentiment_label'] = volume_summary['sentiment_numeric'].map({0:'Fear', 1:'Greed'})
volume_summary
```

```python
buy_sell_summary = merged.groupby(['sentiment_numeric', 'side_numeric'])['profit_flag'].mean().reset_index()
buy_sell_summary['sentiment_label'] = buy_sell_summary['sentiment_numeric'].map({0:'Fear', 1:'Greed'})
buy_sell_summary['trade_type'] = buy_sell_summary['side_numeric'].map({1:'BUY', -1:'SELL'})
buy_sell_summary
```

```python
sns.barplot(x='sentiment_label', y='profit_flag', data=profit_summary)
plt.title("Average Trade Profitability: Fear vs Greed")
plt.ylabel("Proportion of Profitable Trades")
plt.ylim(0,1)
plt.show()
```

```python
sns.barplot(x='sentiment_label', y='Size USD', data=volume_summary)
plt.title("Average Trade Size (USD): Fear vs Greed")
plt.show()
```

```python
sns.barplot(x='trade_type', y='profit_flag', hue='sentiment_label', data=buy_sell_summary)
plt.title("Profitability by Trade Type under Fear vs Greed")
plt.ylabel("Proportion of Profitable Trades")
plt.show()
```
